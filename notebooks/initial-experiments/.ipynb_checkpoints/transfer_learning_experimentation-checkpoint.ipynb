#### Load the data

```python
# Set directories
train_dir = os.path.join('/content/data/regular', 'train')
val_dir = os.path.join('/content/data/regular', 'validation')
test_dir = os.path.join('/content/data/regular', 'test')
```

### Initial Transfer Learning Experiments


#### Model params and fitting

```python
params = {'model':'VGG16',
          'freeze':False,
          'trainable':'last_layer', 
          'dense_layers':0, 
          'dense_hidden_size':8,
          'dropout_rate':0, 
          'l1_reg':0, 
          'l2_reg':0, 
          'optimizer_function':'SGD', 
          'learning_rate':0.001,
          'momentum':0.1,
          'nesterov':True,
          'decay':0,
          'print_model':False}
```

```python
batch_size = 100

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(224, 224),
                                                    batch_size=batch_size,
                                                    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(val_dir,
                                                        target_size=(224, 224),
                                                        batch_size=batch_size,
                                                        class_mode='binary')
```

```python
model = create_embedding(**params)
hist = fit_model(model, train_generator, validation_generator)
```
